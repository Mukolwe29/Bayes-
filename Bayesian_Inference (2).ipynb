
## Discussion of the Conversation between Yueh Han and Akma

The conversation between Yueh Han and Akma revolves around the probability of getting 30 tails in a row when flipping a fair coin. Yueh Han is skeptical, suggesting the outcome is so unlikely that the coins must be tricked or the experiment rigged. Akma counters that while the probability of getting any specific sequence of 30 flips (including 30 tails) is extremely low, all sequences of 30 flips are equally unlikely, so we can't conclude the coin is unfair based solely on this outcome.

#### Analysis:

**Yueh Han's Argument:**

The probability of getting 30 tails in a row with a fair coin is
$$
\left( \frac{1}{2} \right)^{30} \approx 9 \times 10^{-10}
$$
This is indeed a very low probability.

**Akma's Argument:**

Any specific sequence of 30 coin flips has the same probability of occurring, which is
$$
\left( \frac{1}{2} \right)^{30}
$$
The fact that 30 tails in a row seems unlikely is just as unlikely as any other specific sequence of heads and tails.

#### Conclusion:

Akma's perspective is more statistically sound. While getting 30 tails in a row is extremely unlikely, so is any other specific sequence of 30 flips. Without additional evidence, we can't conclude that the coin is unfair based solely on the observed sequence.

### 2. Bayesian Analysis

Given:

- Fair coin: Probability of tails, $P(T|F) = 0.5$
- Double-tailed coin: Probability of tails, $P(T|D) = 1$
- Observed data: 30 tails in a row

We need to find the posterior probability that the coin is fair given the observed data.

Bayes' theorem states:
$$
P(F|D) = \frac{P(D|F)P(F)}{P(D)}
$$

Where:

- $P(D|F)$ is the likelihood of observing the data given the coin is fair.
- $P(F)$ is the prior probability that the coin is fair.
- $P(D)$ is the marginal likelihood of the data.

Calculating $P(D)$:

$$
P(D) = P(D|F)P(F) + P(D|D)P(D)
$$

Where:

$$
P(D|F) = \left( \frac{1}{2} \right)^{30}
$$
$$
P(D|D) = 1
$$

### 3. Posterior Probability Calculation

To determine for which values of the prior \( p \) the posterior probability that the coin is fair is greater than 0.5, we solve:

$$
P(F|D) > 0.5
$$

#### Detailed Calculation:

1. Bayes' theorem:
   $$
   P(F|D) = \frac{P(D|F)P(F)}{P(D)}
   $$

2. Marginal likelihood $P(D)$:
   $$
   P(D) = P(D|F)P(F) + P(D|D)P(D)
   $$
   $$
   P(D) = \left( \frac{1}{2} \right)^{30} P(F) + 1 \cdot P(D)
   $$
   $$
   P(D) = \left( \frac{1}{2} \right)^{30} p + (1 - p)
   $$

3. Posterior probability:
   $$
   P(F|D) = \frac{\left( \frac{1}{2} \right)^{30} p}{\left( \frac{1}{2} \right)^{30} p + (1 - p)}
   $$

4. Setting $P(F|D) > 0.5$:
   $$
   \frac{\left( \frac{1}{2} \right)^{30} p}{\left( \frac{1}{2} \right)^{30} p + (1 - p)} > 0.5
   $$

5. Simplifying the inequality:
   $$
   \left( \frac{1}{2} \right)^{30} p > 0.5 \left( \left( \frac{1}{2} \right)^{30} p + (1 - p) \right)
   $$
   $$
   \left( \frac{1}{2} \right)^{30} p > \left( \frac{1}{2} \right)^{31} p + 0.5 - 0.5p
   $$
   $$
   \left( \frac{1}{2} \right)^{30} p - \left( \frac{1}{2} \right)^{31} p > 0.5 - 0.5p
   $$
   $$
   \left( \frac{1}{2} \right)^{30} p \left( 1 - \frac{1}{2} \right) > 0.5 - 0.5p
   $$
   $$
   \left( \frac{1}{2} \right)^{30} \frac{p}{2} > 0.5 - 0.5p
   $$
   $$
   \left( \frac{1}{2} \right)^{31} p > 0.5 - 0.5p
   $$
   $$
   \left( \frac{1}{2} \right)^{31} p + 0.5p > 0.5
   $$
   $$
   p \left( \left( \frac{1}{2} \right)^{31} + 0.5 \right) > 0.5
   $$
   $$
   p > \frac{0.5}{\left( \frac{1}{2} \right)^{31} + 0.5}
   $$

   Simplifying further:
   $$
   p > \frac{0.5}{\left( \frac{1}{2} \right)^{31} + 0.5}
   $$
   $$
   p > \frac{0.5}{\left( \frac{1}{2} \right)^{31} + \frac{1}{2}}
   $$
   $$
   p > \frac{0.5}{\frac{1}{2^{31}} + \frac{1}{2}}
   $$
   $$
   p > \frac{0.5}{\frac{1}{2^{31}} + \frac{1}{2}}
   $$
   $$
   p > \frac{0.5}{\frac{1 + 2^{30}}{2^{31}}}
   $$
   $$
   p > \frac{0.5 \cdot 2^{31}}{1 + 2^{30}}
   $$
   $$
   p > \frac{2^{30}}{1 + 2^{30}}
   $$

Therefore, the prior  $p$ must be greater than $\frac{2^{30}}{1 + 2^{30}}$ for the posterior probability that the coin is fair to be greater than 0.5.


```python
# Define constants
p_F_given_D = (1/2)**30  # Probability of observing 30 tails in a row given a fair coin
p_D_given_D = 1          # Probability of observing 30 tails in a row given a double-tailed coin

# Function to calculate posterior probability
def posterior_prob(p):
    p_D = p_F_given_D * p + p_D_given_D * (1 - p)
    p_F_given_D_posterior = (p_F_given_D * p) / p_D
    return p_F_given_D_posterior

# Calculate the threshold value for the prior p
threshold = (2**30) / (1 + 2**30)

# Display the result
print(f"The prior probability p must be greater than {threshold:.10f} for the posterior probability that the coin is fair to be greater than 0.5.")

# Example: Check the posterior probability for p slightly above and below the threshold
p_below_threshold = threshold - 0.0000000001
p_above_threshold = threshold + 0.0000000001

posterior_below = posterior_prob(p_below_threshold)
posterior_above = posterior_prob(p_above_threshold)

print(f"Posterior probability for p just below the threshold ({p_below_threshold:.10f}): {posterior_below:.10f}")
print(f"Posterior probability for p just above the threshold ({p_above_threshold:.10f}): {posterior_above:.10f}")

```

```python
# Plotting the posterior probability as a function of the prior p
p_values = np.linspace(0, 1, 10000)
posterior_values = [posterior_prob(p) for p in p_values]

plt.figure(figsize=(10, 6))
plt.plot(p_values, posterior_values, label='Posterior Probability P(F|D)')
plt.axhline(y=0.5, color='r', linestyle='--', label='Posterior Probability = 0.5')
plt.axvline(x=threshold, color='g', linestyle='--', label=f'Prior Threshold = {threshold:.10f}')
plt.xlabel('Prior Probability p')
plt.ylabel('Posterior Probability P(F|D)')
plt.title('Posterior Probability vs. Prior Probability')
plt.legend()
plt.grid(True)
plt.show()
```

## Problem 2: Medical test

### Posterior Probability with Conditionally Independent Test Results

Given:
- $D$: Omer has the disease.
- $T_j$: Omer tests positive on the  $j$-th test.
- $p = P(D)$: Prior probability that Omer has the disease.
- $a_0 = P(T_j \mid D)$: Probability of testing positive given that Omer has the disease.
- $b_0 = P(T_j \mid D^c)$: Probability of testing positive given that Omer does not have the disease.

We need to find the posterior probability that Omer has the disease given that he tests positive on all $n$tests.

Since the test results are conditionally independent given Omer's disease status:
$$
 P(T_1 \cap T_2 \cap \ldots \cap T_n \mid D) = (a_0)^n
$$
$$
  P(T_1 \cap T_2 \cap \ldots \cap T_n \mid D^c) = (b_0)^n
$$

Using Bayes' theorem, the posterior probability is:
$$
P(D \mid T_1 \cap T_2 \cap \ldots \cap T_n) = \frac{P(T_1 \cap T_2 \cap \ldots \cap T_n \mid D) P(D)}{P(T_1 \cap T_2 \cap \ldots \cap T_n)}
$$

Where the denominator is : $$ \( P(T_1 \cap T_2 \cap \ldots \cap T_n) $$

$$ P(T_1 \cap T_2 \cap \ldots \cap T_n) = P(T_1 \cap T_2 \cap \ldots \cap T_n \mid D) P(D) + P(T_1 \cap T_2 \cap \ldots \cap T_n \mid D^c) P(D^c) $$

$$ P(T_1 \cap T_2 \cap \ldots \cap T_n) = (a_0)^n p + (b_0)^n (1 - p) $$

Therefore, the posterior probability becomes:

$$ P(D \mid T_1 \cap T_2 \cap \ldots \cap T_n) = \frac{(a_0)^n p}{(a_0)^n p + (b_0)^n (1 - p)} $$


### Posterior Probability with Genetic Condition

Given:
$G$: Omer has the gene that makes him always test positive.
$P(G) = \frac{1}{2} \)$
$D$ and $G$ are independent.
If $G$, then $P(T_j \mid G) = 1$.
If $G^c$, then $a_1 = P(T_j \mid D, G^c)$ and $b_1 = P(T_j \mid D^c, G^c)$.

We need to find the posterior probability that Omer has the disease given that he tests positive on all $n$ tests, considering the gene condition.

Using Bayes' theorem and considering the gene condition, the posterior probability is:

$$ P(D \mid T_1 \cap T_2 \cap \ldots \cap T_n) = \frac{P(T_1 \cap T_2 \cap \ldots \cap T_n \mid D, G^c) P(D, G^c) + P(T_1 \cap T_2 \cap \ldots \cap T_n \mid D, G) P(D, G)}{P(T_1 \cap T_2 \cap \ldots \cap T_n)} $$

But since $D$ and $G$ are independent, $P(D, G^c) = P(D)$ $P(G^c) = p \cdot \frac{1}{2} \)$, and $P(D, G) = P(D) P(G) = p \cdot \frac{1}{2}$.

Similarly, we can find the joint probability for $D^c$:

$$ P(D^c, G^c) = (1 - p) \cdot \frac{1}{2} $$
$$ P(D^c, G) = (1 - p) \cdot \frac{1}{2} $$

Given G,
$$ P(T_1 \cap T_2 \cap \ldots \cap T_n \mid G) = 1 $$.
Given $$G^c $$, the tests are conditionally independent as described in part 1.

The denominator becomes:

$$ P(T_1 \cap T_2 \cap \ldots \cap T_n) = P(T_1 \cap T_2 \cap \ldots \cap T_n \mid D, G^c) P(D, G^c) + P(T_1 \cap T_2 \cap \ldots \cap T_n \mid D^c, G^c) P(D^c, G^c) + P(T_1 \cap T_2 \cap \ldots \cap T_n \mid G) P(G) $$

$$ P(T_1 \cap T_2 \cap \ldots \cap T_n) = (a_1)^n \cdot \frac{p}{2} + (b_1)^n \cdot \frac{1 - p}{2} + 1 \cdot \frac{1}{2} $$

Therefore, the posterior probability of having the disease is:

$$ P(D \mid T_1 \cap T_2 \cap \ldots \cap T_n) = \frac{(a_1)^n \cdot \frac{p}{2}}{(a_1)^n \cdot \frac{p}{2} + (b_1)^n \cdot \frac{1 - p}{2} + \frac{1}{2}} $$

### Posterior Probability of Having the Gene

Using the same setup as in part 2, we need to find the posterior probability that Omer has the gene given that he tests positive on all $n$ tests.

The posterior probability is:

$$ P(G \mid T_1 \cap T_2 \cap \ldots \cap T_n) = \frac{P(T_1 \cap T_2 \cap \ldots \cap T_n \mid G) P(G)}{P(T_1 \cap T_2 \cap \ldots \cap T_n)} $$

$$ P(G \mid T_1 \cap T_2 \cap \ldots \cap T_n) = \frac{1 \cdot \frac{1}{2}}{(a_1)^n \cdot \frac{p}{2} + (b_1)^n \cdot \frac{1 - p}{2} + \frac{1}{2}} $$



```python
# Given parameters
n = 30  # number of tests
p = 0.1  # prior probability of having the disease
a0 = 0.99  # P(Tj | D) for part 1
b0 = 0.01  # P(Tj | Dc) for part 1

# Part 1: Posterior probability given conditionally independent test results
posterior_disease_part1 = (a0**n * p) / (a0**n * p + b0**n * (1 - p))

# Part 2: Considering the gene condition
a1 = 0.99  # P(Tj | D, Gc)
b1 = 0.01  # P(Tj | Dc, Gc)

denominator = (a1**n * p / 2) + (b1**n * (1 - p) / 2) + 0.5
posterior_disease_part2 = (a1**n * p / 2) / denominator

# Part 3: Posterior probability of having the gene
posterior_gene = 0.5 / denominator

print(f"Posterior probability of disease (part 1): {posterior_disease_part1:.6f}")
print(f"Posterior probability of disease (part 2): {posterior_disease_part2:.6f}")
print(f"Posterior probability of having the gene (part 3): {posterior_gene:.6f}")
```

## Problem 3. Proofreading (Distrubutions, ModelSelection)

Given:
## 1. Distribution of \(X\)

Given:

- $n$: Number of typos in the book.

- $p_1$: Probability that Sho finds a typo.

- $p_2$: Probability that Haruna finds a typo.

- $X_1$: Number of typos caught by Sho.

- $X_2$: Number of typos caught by Haruna.

- $X$: Number of typos caught by at least one of the proofreaders.

The distribution of $X$ can be derived by considering the probability of a typo being found by at least one of the proofreaders.

The probability that a typo is found by Sho or Haruna (or both) is given by:
$$
P(T) = P(\text{Sho finds typo}) + P(\text{Haruna finds typo}) - P(\text{Both find typo}) = p_1 + p_2 - p_1 p_2
$$

The number of typos found by at least one of the proofreaders $X$ follows a binomial distribution:
$$
 X \sim \text{Binomial}(n, p_1 + p_2 - p_1 p_2)
$$

## 2. Conditional Distribution of $X_1$ given $X_1 + X_2 = t$

Given $p_1 = p_2 = p$, and we know $X_1 + X_2 = t$.

The number of typos found by Sho and Haruna is given by the sum

$X_1 + X_2$,

and we know:

$$
 X_1 + X_2 = X + \text{number of typos found by both Sho and Haruna}
$$

This means that if we condition on $X_1 + X_2 = t$, we are looking at the typos found by Sho plus the typos found by Haruna, double-counting the ones found by both.

Under this condition, we need to find the distribution of $X_1$.

Since $X_1$ and $X_2$ are binomially distributed, and given $X_1 + X_2 = t$, the conditional distribution of $X_1$ is a hypergeometric distribution:

$$
 X_1 \mid (X_1 + X_2 = t) \sim \text{Hypergeometric}(n, p, t)
$$


### 3. Model Selection to Decide if Haruna is Better

To decide if Haruna is better than Sho at finding typos, we can use a frequentist hypothesis testing approach. Specifically, we can simulate the distribution of the number of typos found by Sho and Haruna under the null hypothesis that both proofreaders have the same probability of finding typos.

Let's denote:

- $H_0$: Sho and Haruna have the same probability of finding typos $(p_1 = p_2)$.
- $H_1$: Haruna is better than Sho at finding typos $(p_2 > p_1)$.

### Simulation for Hypothesis Testing

We'll use the following steps:

1. Estimate the total number of typos \(n\) based on the average typo rate.
2. Simulate the number of typos found by Sho and Haruna under the null hypothesis.
3. Compare the observed number of typos found by Sho and Haruna to the simulated distribution.



```python
# Given data
words = 100000
typo_rate = 1 / 300
n = int(words * typo_rate)  # Estimate of total typos
p1 = p2 = 299 / n  # Sho's typo finding rate as the baseline (under null hypothesis)
observed_sho = 299
observed_haruna = 314

# Simulation parameters
num_simulations = 10000
simulated_sho = np.zeros(num_simulations)
simulated_haruna = np.zeros(num_simulations)

for i in range(num_simulations):
    typos = np.random.binomial(n, p1, size=2)
    simulated_sho[i] = typos[0]
    simulated_haruna[i] = typos[1]

# Calculate p-value for Haruna being better than Sho
p_value = np.mean(simulated_haruna >= observed_haruna)

print(f"Estimated total typos: {n}")
print(f"Simulated p-value for Haruna being better than Sho: {p_value:.4f}")
```

### 4. Estimating the Number of Typos Left Using Bayesian Parameter Estimation

In this part, we aim to estimate the probability that Sho and Haruna find typos $(p_1)$ and $(p_2)$, and the total number of typos $(t)$, given the data about the typos they found independently and jointly.

Given:
- Sho found 299 typos.
- Haruna found 314 typos.
- Both found 285 typos.
- Sho found 14 typos that Haruna did not find.
- Haruna found 29 typos that Sho did not find.

To estimate the parameters using Bayesian inference, we need to:
1. Define prior distributions for $(p_1)$, $(p_2)$, and $(t)$.
2. Use the observed data to update these priors and get posterior distributions.

We'll use a simple Monte Carlo simulation to approximate the posterior distributions.

```python
# Given data
found_by_sho = 299
found_by_haruna = 314
found_by_both = 285
found_by_sho_only = 14
found_by_haruna_only = 29

# Prior distributions (assuming uniform priors for simplicity)
prior_p1 = stats.uniform(0, 1)
prior_p2 = stats.uniform(0, 1)
prior_t = stats.randint(0, 1000)  # Arbitrary upper bound

# Number of samples for Monte Carlo simulation
num_samples = 10000
samples_p1 = prior_p1.rvs(size=num_samples)
samples_p2 = prior_p2.rvs(size=num_samples)
samples_t = prior_t.rvs(size=num_samples)

# Posterior sampling
posterior_samples = []

for i in range(num_samples):
    p1 = samples_p1[i]
    p2 = samples_p2[i]
    t = samples_t[i]

    # Likelihood of the observed data given the parameters
    likelihood_sho = stats.binom.pmf(found_by_sho, t, p1)
    likelihood_haruna = stats.binom.pmf(found_by_haruna, t, p2)
    likelihood_both = stats.binom.pmf(found_by_both, t, p1 * p2)

    # Posterior probability (unnormalized)
    posterior_prob = likelihood_sho * likelihood_haruna * likelihood_both

    posterior_samples.append((p1, p2, t, posterior_prob))

# Normalize posterior probabilities
posterior_samples = np.array(posterior_samples)
posterior_probs = posterior_samples[:, 3]
posterior_probs /= np.sum(posterior_probs)

# Extract posterior estimates
posterior_p1 = np.sum(posterior_samples[:, 0] * posterior_probs)
posterior_p2 = np.sum(posterior_samples[:, 1] * posterior_probs)
posterior_t = np.sum(posterior_samples[:, 2] * posterior_probs)

# Number of typos left in the book
typos_left = posterior_t - (found_by_sho + found_by_haruna - found_by_both)

print(f"Posterior estimate of p1: {posterior_p1:.4f}")
print(f"Posterior estimate of p2: {posterior_p2:.4f}")
print(f"Posterior estimate of total typos (t): {posterior_t:.4f}")
print(f"Estimated number of typos left in the book: {typos_left:.4f}")
```

## Problem 4. Randomized response surveys(ParameterEstimation)
### Estimating the Proportion of Participants Who Use Drugs

#### 1. Finding $y$ in Terms of $d$ and $p$
Let $d$ be the probability that a participant has used illegal drugs, and $p$ be the proportion of slips that say "I have used illegal drugs".

When a participant draws a slip, they have two scenarios:

1. The slip says "I have used illegal drugs" (with probability $p$, and the participant says "yes" with probability \( d \).
2. The slip says "I have not used illegal drugs" (with probability $1 - p$, and the participant says "yes" with probability $0$ (since the slip says "I have not used illegal drugs").

Thus, the probability $y$ that a participant says "yes" can be expressed as:
$$
y = p \cdot d + (1 - p) \cdot 0 = p \cdot d
$$

#### 2. Worst Possible Choice of $p$)
The worst possible choice of $p$ would be one that provides the least amount of information about the true proportion $d$. If $p$ is too high or too low, the responses might not accurately reflect the participants' drug use.

Consider two extreme cases:
- If $p = 1$, everyone is answering based on whether they have used drugs, which might lead to biased responses due to the sensitive nature of the question.
- If $p = 0$, no one answers based on drug use, providing no useful information about $d$.

The optimal choice is typically a balanced $p$ to maximize the information gain while reducing the bias due to the sensitive nature of the question. A common choice is $p = 0.5$.

#### 3. Finding $d$ in Terms of $y$ and $p$
Now, with the new setup where slips say "I have used illegal drugs" with probability $p$ and "I was born in winter" with probability  $1 - p$:

Let $q$ be the probability that a participant was born in winter, which is given as $q = \frac{1}{4}$.

When a participant draws a slip:
- With probability $p$, the slip says "I have used illegal drugs" and they say "yes" with probability $d$.
- With probability $1 - p$, the slip says "I was born in winter" and they say "yes" with probability $q$.

Thus, the probability $y$ that a participant says "yes" is:
$$
y = p \cdot d + (1 - p) \cdot q
$$
$$
y = p \cdot d + (1 - p) \cdot \frac{1}{4}
$$

Solving for $d$:
$$
y = p \cdot d + \frac{1 - p}{4}
$$
$$
y - \frac{1 - p}{4} = p \cdot d
$$
$$
d = \frac{y - \frac{1 - p}{4}}{p}
$$
$$
d = \frac{y - \frac{1}{4} + \frac{p}{4}}{p}
$$
$$
d = \frac{4y - 1 + p}{4p}
$$

#### 4. Estimating the Proportion of Participants Who Use Drugs
Given:
- $p = \frac{1}{2}$
- 100 participants
- 20 "yes" responses

Using the formula from part 3:
$$
d = \frac{y - \frac{1 - p}{4}}{p}
$$
$$
d = \frac{\frac{20}{100} - \frac{1 - \frac{1}{2}}{4}}{\frac{1}{2}}
$$
$$
d = \frac{0.2 - \frac{1}{8}}{0.5}
$$
$$
d = \frac{0.2 - 0.125}{0.5}
$$
$$
d = \frac{0.075}{0.5}
$$
$$
d = 0.15
$$

So, the point estimate for the proportion of participants who use drugs is $0.15$.

#### Bayesian Posterior Simulation
To perform a Bayesian posterior simulation, we need to generate a posterior distribution for  $d$. Let's assume a uniform prior for  $d$ between 0 and 1.



**Bayesian Posterior Simulation**

```python
# Parameters
p = 0.5
num_participants = 100
num_yes_responses = 20
prior_d = np.random.uniform(0, 1, 10000)  # Uniform prior

# Simulation
posterior_samples = []
for d in prior_d:
    y_prob = p * d + (1 - p) * 0.25
    likelihood = y_prob ** num_yes_responses * (1 - y_prob) ** (num_participants - num_yes_responses)
    posterior_samples.append((d, likelihood))

# Normalize the posterior samples
posterior_samples = np.array(posterior_samples)
posterior_samples[:, 1] /= np.sum(posterior_samples[:, 1])

# Plot the posterior distribution
plt.hist(posterior_samples[:, 0], weights=posterior_samples[:, 1], bins=50, density=True)
plt.xlabel('Proportion of participants who use drugs (d)')
plt.ylabel('Density')
plt.title('Posterior distribution of d')
plt.show()

# Calculate mean of the posterior
posterior_mean_d = np.sum(posterior_samples[:, 0] * posterior_samples[:, 1])
print(f'Posterior mean of d: {posterior_mean_d:.4f}')
```
